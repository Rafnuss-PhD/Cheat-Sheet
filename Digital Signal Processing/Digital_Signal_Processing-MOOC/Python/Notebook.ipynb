## 2.3 The Karplus-Strong Algo

$y[n]=\alpha y[n-M]+x[n]$

<img  src="files/karplus-strong.jpg">

$x[n]$ is the input signal, $M$ is the delay, and $\alpha$ is the decay. We assume that the value of the delay M is equal to the length of the input signal x. In words, it is like filling the delay buffer and then recursively going over it. Additionally, we set the length of the output to be a multiple of the delay M (i.e., of the length of the input signal). This is controlled by the argument D, so that the output length is $D∗M$.

```python
def ks(x, alpha, D) :
    import numpy as np
    #   Length of the output signal must be larger than the length of the input signal,
    #   that is, D must be larger than 1 
    if D < 1:
        print('Duration D must be greater than 1')
        
    #   Make sure the input is a row-vector
    if x.ndim != 1:
        print('The array entered is of the wrong size')
        return None
    #   Number of input samples
    M = len(x)
    
    #   Create a vector of the powers of alpha, [alpha^0 alpha^1 ....]
    a = np.ones((1,D)) * alpha
    b = np.arange(D)
    alphaVector = pow(a,b)
    
    #Create a matrix with M columns, each being the vector of the powers of alpha
    alphaMatrix = np.eye(D,M) 
    for index in range(M):
        alphaMatrix[:,index] = alphaVector
        
    #Create a matrix with D rows filled by the input signal x  
    xMatrix = np.tile(x,(D,1))
    
    #Multipliy the two, so we can read it out
    #column-by-column
    yMatrix = alphaMatrix * xMatrix
    
    #Read out the output column by columnn
    y = yMatrix.flatten()
        
    return y
```

```python
# Parameters:
#
# - Fs       : sampling frequency
# - F0       : frequency of the notes forming chord
# - gain     : gains of individual notes in the chord
# - duration : duration of the chord in second
# - alpha    : attenuation in KS algorithm

Fs = 48000

import numpy as np
# D2, D3, F3, G3, F4, A4, C5, G5
F0 = 440*np.array([pow(2,(-31./12.)), pow(2,(-19./12.)), pow(2,(-16./12.)), pow(2,(-14./12.)), pow(2,(-4./12.)), 1, pow(2,(3./12.)), pow(2,(10./12.))])
gain = np.array([1.2, 3.0, 1.0, 2.2, 1.0, 1.0, 1.0, 3.5])
duration = 4
alpha = 0.9785

# Number of samples in the chord
nbsample_chord = Fs * duration

# This is used to correct alpha later, so that all the notes decay together
# (with the same decay rate)
first_duration = np.ceil(float(nbsample_chord)/round(float(Fs)/float(F0[0])))

# Initialization
chord = np.zeros(nbsample_chord)

for i in range(len(F0)):
    
    # Get M and duration parameter
    current_M = round(float(Fs)/float(F0[i]));
    current_duration = np.ceil(float(nbsample_chord)/float(current_M))
    
    # Correct current alpha so that all the notes decay together (with the
    # same decay rate)
    current_alpha = pow(alpha,(float(first_duration)/float(current_duration)))
    
    # Let Paul's high D on the bass ring a bit longer
    if i == 1:
        current_alpha = pow(current_alpha,8)
    
    # Generate input and output of KS algorithm
    x = np.random.rand(current_M)
    y = ks(x, current_alpha, current_duration)
    y = y[0:nbsample_chord]
    
    # Construct the chord by adding the generated note (with the
    # appropriate gain)
    chord = chord + gain[i] * y

import numpy as np
from scipy.io.wavfile import write

data = chord
scaled = np.int16(data/np.max(np.abs(data)) * 32767)

write('hard_days.wav', 44100, scaled)
```

## 3.2 Hilbert Space, proterties and bases

Definition:
A Hilbert space H is a real or complex inner product space that is also a complete metric space with respect to the distance function induced by the inner product.

In mathematics, the dot product, or scalar product (or sometimes inner product in the context of Euclidean space), is an algebraic operation that takes two equal-length sequences of numbers (usually coordinate vectors) and returns a single number. Properties includes : cummutative, distributive, biliniear, scalar multiplication, orthogonal, no cancellation and product rules.

* In algebric definition: $\mathbf{A}\cdot \mathbf{B} = \sum_{i=1}^n A_iB_i = A_1B_1 + A_2B_2 + \cdots + A_nB_n$
* In geometric difinition: $\mathbf A\cdot\mathbf B = \|\mathbf A\|\,\|\mathbf B\|\cos\theta$
* For function : $\langle u , v \rangle = \int_a^b u(x)v(x)dx $
* For signals: $\langle x , y \rangle =\sum_{n=-\infty}^{\infty} x[n]y[n]$. But it require vector to be square-summable: $\sum |x[n]|^2 < \infty$.

By-product:
* Norm: $\|x\| = \sqrt{\langle x , x \rangle}$
* Distence: $d(x,y)= \| x-y \|$

With $w_k[n]=e^{j\frac{2\pi}{N}nk}$. Be a set of K vector from H : $W={\boldsymbol{w}^{(k)}}_{k=0,...K-1}$. W is a basis of K if we can write for all $x \in H$, $x=\sum \alpha_k \boldsymbol{w}^k$ with $\alpha_k \in \mathbb{C}$

This also imply linear independance : $\sum \alpha_k \boldsymbol{w}^k = 0 \equiv \alpha_k=0, k=0,...K-1$

Hilbert space is a generalisation of Euclidean space. A Hilbert space is an abstract vector space possessing the structure of an inner product that allows length and angle to be measured.

$\boldsymbol{x}$ from an Hilbert space can be expend in an orthogonal space:

### Approximation by projection

How to approximate a vector $\boldsymbol{x} \in V$ in a subspace $S \subseteq V$ ? *Least-Square estimate:*
$$\boldsymbol{\hat{x}}=\sum_{k=0}^{K-1} \langle \boldsymbol{s}^{(k)}, \boldsymbol{x}\rangle \boldsymbol{s}^{(k)}$$
Orthogoanl projection is the best approximation !
The error ($\boldsymbol{x}-\boldsymbol{\hat{x}}$) is orthogonal to the approximation: $$\langle \boldsymbol{x}-\boldsymbol{\hat{x}}, \boldsymbol{\hat{x}} \rangle$$

### Exemples

The goal is to approximate $x=\sin(t)$ over $P_3[-1,1]$ (polynomial up to degree 3).
Steps:
1. Build orthogonal basis from the naive basis
2. Project x over the orthonormal basis
3. Compute approximation error
4. compare error to Taylor expension

1. Build orthogonal basis from the naive basis

Using the Gram-Schmidt algorithm, we can get vector space of the orthoganl basis from the naive basis ($\boldsymbol{s}^{(k)}=t^k$) (not orthogonal)
$$\boldsymbol{u}^{(0)}=\sqrt{1/2}, \quad \boldsymbol{u}^{(1)}=\sqrt{3/2}t, \quad \boldsymbol{u}^{(2)}=\sqrt{5/8}(3t^2-1),...$$
such that now, $\langle \boldsymbol{u}^{(i)}, \boldsymbol{u}^{(j)} \rangle=0$ for $i \neq j$
<img src="files/exemple.jpg">

2. Project x over the orthonormal basis

We want to find $\alpha_k$:
$$\boldsymbol{\hat{x}}=\sum_{k=0}^{K-1} \langle \boldsymbol{u}^{(k)}, \boldsymbol{x}\rangle \boldsymbol{u}^{(k)}$$
where:
$$\langle \boldsymbol{u}^{(k)}, \boldsymbol{x}\rangle= \int_{-1}^1 u_k(t) \sin t dt$$
This end up to give:
$$\alpha_0=0, \quad \alpha_1\approx 0.74, \quad\alpha_2=0$$
Therefore,
$$\sin t = \alpha_1 \boldsymbol{u}^{(1)}\approx 0.9035 t$$

3. Compute approximation error

4. compare error to Taylor expension

$$\sin t \approx t$$

```python
t=np.linspace(-1,1,1000)
plt.plot(t,np.sin(t),label='sin(t)')
plt.plot(t,t,label='Taylor (t)')
plt.plot(t,0.9035*t,label='(0.9035 t)')
plt.legend(); plt.show()
plt.plot(t,np.sin(t)-t,label='sin(t)-t')
plt.plot(t,np.sin(t)-0.9035*t,label='sin(t)- 0.9035 t')
plt.legend(); plt.show()
```

### The Fourier Basis

The set of N signals in $\mathbb{C}^N$. (N signals of N dimensions) defined by N vectors:
$$w_k[n]=e^{j\frac{2 \pi}{N}nk}, \quad n,k=0,1,...,N-1$$
is an orthogonal basis in $\mathbb{C}^N$ because:
$$\langle \boldsymbol{w}^{(k)} , \boldsymbol{w}^{(h)} \rangle =\sum_{n=0}^{N-1} e^{j\frac{2 \pi}{N}n(h-k)}=\frac{1-e^{j2 \pi (h-k)}}{1-e^{j \frac{2 \pi}{N} (h-k)}}=0$$
But not orthonormal: $\langle \boldsymbol{w}^{(k)} , \boldsymbol{w}^{(k)} \rangle = N$. Noramlization factor is $1/\sqrt{N}$

```python
N=32;
w=np.empty([N,N],dtype=np.complex128)
for k in range(N):
    for n in range(N):
        w[k][n]=np.exp(1j*2*np.pi/N*n*k)
        plt.polar([0,np.angle(w[k][n])],[0,np.abs(w[k][n])],marker='o')
plt.show()

plt.plot(w[:][0:4].real.T); plt.show()
plt.plot(w[:][0:4].imag.T); plt.show()
```

## 4.2: The Discrete Fourier Transform (DFT)

### The Discrete Foourier Transform (DFT)

1.  FT as vector basis transfromation
How to express a vector $\boldsymbol{x}$ in the fourrier basis ?
$$\boldsymbol{X}_k = \langle \boldsymbol{w}^{(k)} , \boldsymbol{x} \rangle = \boldsymbol{W}\boldsymbol{x}$$
where $W[n,m]=e^{j\frac{2 \pi}{N}nm}$

2. FT as a canonical basis transformation
Usually, we express a value which change in time as a vector where each element correspond at a certain time : $\boldsymbol{x}=\sum x_k e^k$. Where $e^k$ is a the canonical vector which is delta function $\delta(n-k)$ ($e^k=1$ for $k=n$, else $e^k=0$). Fourrier transform is not using this canonical basis but sinus basis.

3. Exact equation
$$X[k] = \sum_{n=0}^{N-1} x[n] e^{- j\frac{2 \pi}{N}nk}$$

When we talk about time, this is not the time is second or whatever but the index that move over the range of x.

The reverse formula (synthesis formula):
$$x[n] = \frac{1}{N} \sum_{k=0}^{N-1} X[k] e^{j\frac{2\pi}{N}nk}$$

### DFT exemple

Be $x[n] \in \mathbb{C}^64$ so that $w_k[n]=e^{j\frac{2\pi}{64}nk}$
$$x[n]=3\cos \left(\frac{2\pi}{16}n+\frac{\pi}{3} \right)= \frac{3}{2}\left( e^{j\frac{2\pi}{64}4n}e^{j\frac{\pi}{3}} + e^{-j\frac{2\pi}{64}4n}e^{-j\frac{\pi}{3}} \right)= \frac{3}{2}\left( e^{j\frac{\pi}{3}} w_4[n]+ e^{-j\frac{\pi}{3}}w_{60}[n] \right)$$
$$X[k]= \langle w_k[n] , x[n] \rangle = \frac{3}{2} \langle w_k[n] , e^{j\frac{\pi}{3}} w_4[n] \rangle     +   \frac{3}{2} \langle w_k[n] , e^{-j\frac{\pi}{3}}w_{60}[n] \rangle= \left\{ \begin{array}{cc} 96e^{j\frac{\pi}{3}} & \text{for k=4} \\ 96e^{-j\frac{\pi}{3}} & \text{for k=60}\\ 0 & \text{otherwise} \end{array} \right.$$

```python
N=64;
n=np.linspace(0,N-1,N);
x=3*np.cos(2*np.pi/16*n+np.pi/3)
X=np.fft.fft(x)
X_ana=np.zeros([N],dtype="complex128")
X_ana[4]=96*np.exp(1j*np.pi/3)
X_ana[60]=96*np.exp(-1j*np.pi/3)
plt.stem(x,label="x");plt.legend();plt.show()
plt.stem(X.real,label="Re(X)");plt.stem(X_ana.real,label="Re(X_ana)",markerfmt='ro');plt.legend();plt.show()
plt.stem(X.imag,label="Im(X)");plt.stem(X_ana.imag,label="Im(X_ana)",markerfmt='ro');plt.legend();plt.show()
```

Be $x[n] \in \mathbb{C}^N$ so that $x[n]=\sum_{h=0}^{M-1} \delta [n-h]$.
$$ X[k]=\sum_{n=0}^{N-1}x[n]e^{-j\frac{2\pi}{N}nk}=\sum_{n=0}^{M-1}e^{-j\frac{2\pi}{N}nk}==\frac{1-e^{j \frac{2 \pi}{N}kM}}{1-e^{j \frac{2 \pi}{N} k}}=\frac{\sin\left( \frac{\pi}{N}Mk\right)}{\sin\left( \frac{\pi}{N}k\right)} e^{-j \frac{\pi}{N} (M-1)k} $$

```python
N=64; M=N/16;
n=np.linspace(0,N-1,N);
x=np.concatenate((np.ones([M,1]), np.zeros([N-M,1])),axis=0)
X=np.fft.fft(x.T)
X_ana=np.sin(np.pi/N*M*n)/np.sin(np.pi/N*n) * np.exp(-1j*np.pi/N*(M-1)*n)
plt.stem(x,label="x");plt.legend();plt.show()
plt.stem(X.T.real,label="Re(X)");plt.stem(X_ana.real,label="Re(X_ana)",markerfmt='ro');plt.legend();plt.show()
plt.stem(X.T.imag,label="Im(X)");plt.stem(X_ana.imag,label="Im(X_ana)",markerfmt='ro');plt.legend();plt.show()
```

```python
def dftmatrix(N):
    '''construct DFT matrix'''
    import numpy as np
    
    # create a 1xN matrix containing indices 0 to N-1
    a = np.expand_dims(np.arange(N), 0)
    
    # take advantage of numpy broadcasting to create the matrix
    WN = np.exp(-2j*np.pi*a.T*a/N)
    
    return WN
```

## 4.5 DTFT properties

```python
def dtftmatrix(N):
    '''construct DTFT matrix'''
    import numpy as np
    
    # create a 1xN matrix containing indices 0 to N-1
    a = np.expand_dims(np.arange(N), 0)
    
    # take advantage of numpy broadcasting to create the matrix
    WN = np.exp(-2j*np.pi*a.T*a/N)
    
    return WN
```
